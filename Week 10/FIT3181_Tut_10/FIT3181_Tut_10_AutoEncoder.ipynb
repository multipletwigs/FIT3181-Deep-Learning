## <span style="color:#0b486b">I. Introduction of AutoEncoder</span> ##

### <span style="color:#0b486b">I.1. Overview of Auto-encoders</span> ###
Autoencoders are a specific type of deep neural networks where the input is the same as the output. They encode the input into a lower-dimensional code and then reconstruct the output from this representation. The code is a compact “summary” or “compression” of the input, also called the `latent-space representation`.

An autoencoder consists of 3 components: `encoder`, `latent space` and `decoder`. The encoder compresses the input and produces the code (latent representation), the decoder then reconstructs the input only using this code.

<img src='./images/auto_encoder_general.png' width='400' />

Auto-encoders usually have the following architecture. The input $x$ is fed to the encoder to obtain the code $z = f_{enc}(x)$. The code $z$ is fed to the decoder to yield the reconstruction $\tilde{x} = g_{dec}(z)$. To train an auto-encoder, we minimize the reconstruction error between $x$ and $\tilde{x}$ with the hope to reconstruct $x$ as much as possible so that the latent code $z$ preserves the crucial information of $x$.
- $\min \mathbf{E}_x[d(x, \tilde{x})]$ where $d(x, \tilde{x})$ represents a distance between $x$ and $\tilde{x}$.

<img src='./images/auto_encoder_architecture.png' width='400' align='center'/>


In addition, if you want to decrease the number of weight matrices learnt, you can tie the weight matrices.
- $(W_1^{enc})^T= W_1^{dec}, (W_2^{enc})^T= W_2^{dec}$, and $(W_3^{enc})^T= W_3^{dec}$.

### <span style="color:#0b486b">I.2. Denoising Auto-encoder</span> ###
To allow auto-encoder to denoise noisy inputs, in `Denoising Auto-encoder` (DAE), we add small noises to inputs and try to reconstruct original inputs.


<img src='./images/denoising_AE.png' width='500' align='center'/>

### <span style="color:#0b486b">I.3. Sparse Auto-encoder</span> ###

We encourage the sparsity of the latent code (representation) $z$ by adding some regularizations:
- L1: $\Omega(z)=\left\Vert z\right\Vert _{1}=\sum_{i=1}^{m}|z_{i}|$.
- L2: $\Omega(z)=\frac{1}{2}\left\Vert z\right\Vert _{2}^{2}=\sum_{i=1}^{m}z_{i}^{2}$.
- Cross-entropy: $\Omega(z)=\frac{1}{m}\sum_{i=1}^{m}CE([z_{i},1-z_{i}],[a,1-a])=\frac{1}{m}\sum_{i=1}^{m}[z_{i}\log(a)+(1-z_{i})\log(1-a)]$ where $0<a<1$ is a tiny number.

### <span style="color:#0b486b">I.4. Variational Auto-encoder</span> ###

Variational Auto-encoder (VAE) is a probabilistic version of auto-encoder. Let $p(z)$ (e.g., $N(\mathbf{0}, \mathbf{I})$) be a prior distribution over the latent space.

In VAE, the decoder is probabilistic and represented as a Gaussian distribution $p(x|z) = N(x|\mu_{dec}(z), \sigma^2 \mathbf{I})$. Similarly, the encoder is probabilistic and represented as $q(z|x) = N(z|\mu_{enc}(x), \Sigma_{enc}(x))$.

The objective function to train a VAE is as follows:
- $\min \lambda \mathbf{E}_x[\mathbf{E}_{\epsilon\sim N(0,I)}[d(x, \tilde{x}]] + \mathbf{E}_x[D_{KL}(N(\mu_{enc}(x), \Sigma_{enc}(x))||N(\mathbf{0}, \mathbf{I})]$ where $\epsilon\sim N(0,I)$, $z = \mu(x) + \Sigma(x)^{1/2}\epsilon$, and $\tilde{x} = \mu_{dec}(z)$.
- The first term is known as `reconstruction term` and the second term is known as `regularization term`.

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import math
```

We download the dataset and split it into train/valid/test sets.

### <span style="color:#0b486b">II.2. Declare abstract Auto-Encoder class</span> ###

Here we declare the `GeneralAE` class which is an abstract class for a general auto-encoder serving as a parents class for other classes. There are some attributes (instance variables) and functions (methods) you need to pay attention to:
- $encode$: used to encode input data to latent codes.
- $decode$: used to decode latent codes to reconstructed data.
- $reconstruct$: used to compute the reconstructions of given input data.
- $show\_reconstructions$: used to visualize both original input data and their reconstructions.

```python
class GeneralAE:
    def __init__(self, optimizer = keras.optimizers.SGD(lr=0.1)):
        self.encoder = None
        self.decoder = None
        self.auto_encoder = None
        self.optimizer = optimizer
    
    @staticmethod
    def rounded_accuracy(y_true, y_pred):
        return keras.metrics.binary_accuracy(tf.round(y_true), tf.round(y_pred))
    
    def encode(self, X=None):
        return self.encoder.predict(X)
    
    def decode(self, h=None):
        return self.decoder.predict(h)
    
    def reconstruct(self, X=None):
        self.auto_encoder.predict(X)
    
    def show_reconstructions(self, X= None, n_cols = 5):
        reconstructions = self.auto_encoder.predict(X)
        n_images = len(X)
        n_rows = math.ceil(n_images/n_cols)
        fig = plt.figure(figsize=(2*n_cols*1.5, n_rows*1.5))
        plt.axis("off")
        for i in range(n_images):
            plt.subplot(n_rows, 2*n_cols, 2*i+1)
            plt.imshow(X[i], cmap="gray")
            plt.xlabel("real")
            plt.xticks([])
            plt.yticks([])
            plt.grid(False)
            plt.subplot(n_rows, 2*n_cols, 2*i+2)
            plt.imshow(reconstructions[i], cmap="gray")
            plt.xlabel("reconstruct")
            plt.xticks([])
            plt.yticks([])
            plt.grid(False)
            
    def build(self):
        pass
    
    def train(self, *args, **kwargs):
        self.auto_encoder.fit(*args, **kwargs)
```

### <span style="color:#0b486b">II.3. Implement a standard AutoEncoder</span> ###

#### <span style="color:#0b486b">II.3.1. Fully-connected Standard AutoEncoder </span> ####

This shows the class `StandardAE` which implements the standard AE in which we only minimize the reconstruction loss. Note that the reconstruction loss used in this implementation is binary cross-entropy loss because we scale both input and output pixel intensity to $[0;1]$.

```python
class StandardAE(GeneralAE):
    def __init__(self, optimizer = keras.optimizers.SGD(lr=0.1)):
        super(StandardAE, self).__init__(optimizer)
    
    def build(self):
        self.encoder = keras.models.Sequential([keras.layers.Flatten(input_shape=[28, 28]),
                                                keras.layers.Dense(100, activation="selu"),
                                                keras.layers.Dense(30, activation="selu")])
        
        self.decoder = keras.models.Sequential([keras.layers.Dense(100, activation="selu", input_shape=[30]),
                                                                  keras.layers.Dense(28 * 28, activation="sigmoid"),
                                                                  keras.layers.Reshape([28, 28])])
        
        self.auto_encoder = keras.models.Sequential([self.encoder, self.decoder])
        self.auto_encoder.compile(loss="binary_crossentropy", optimizer=self.optimizer, metrics=[GeneralAE.rounded_accuracy])
```

We create an instance of `StandardAE` and start training.

```python
standard_ae = StandardAE()
standard_ae.build()
standard_ae.train(X_train, X_train, epochs=20, validation_data=(X_valid, X_valid))
```

We show the reconstruction of some images in the valid set.

```python
standard_ae.show_reconstructions(X_valid[:20], n_cols=5)
```

### <span style="color:#0b486b">II.3.2. Convolutional standard AutoEncoder</span> ###

Below is the class `Conv2D_StandardAE` which uses Conv2D layers instead of dense layers as in the previous class. We hence can improve the quality of reconstructed images. Note that in this implementation, we are using `keras.layers.Conv2DTranspose` which will be introduced in detail in the next lecture. For now, you can refer to [here](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Conv2DTranspose) and [here](https://www.machinecurve.com/index.php/2019/12/10/conv2dtranspose-using-2d-transposed-convolutions-with-keras/) for your own investigation. 

```python
class Conv2D_StandardAE(GeneralAE):
    def __init__(self, optimizer = keras.optimizers.SGD(lr=0.1)):
        super(Conv2D_StandardAE, self).__init__(optimizer)
    
    def build(self):
        self.encoder = keras.models.Sequential([keras.layers.Reshape([28, 28, 1], input_shape=[28, 28]),
                                                keras.layers.Conv2D(16, kernel_size=3, padding="SAME", activation="selu"),
                                                keras.layers.MaxPool2D(pool_size=2),
                                                keras.layers.Conv2D(32, kernel_size=3, padding="SAME", activation="selu"),
                                                keras.layers.MaxPool2D(pool_size=2),
                                                keras.layers.Conv2D(64, kernel_size=3, padding="SAME", activation="selu"),
                                                keras.layers.MaxPool2D(pool_size=2)])
        self.decoder = keras.models.Sequential([keras.layers.Conv2DTranspose(32, kernel_size=3, strides=2, 
                                                                         padding="VALID", activation="selu",input_shape=[3, 3, 64]),
                                                keras.layers.Conv2DTranspose(16, kernel_size=3, strides=2, padding="SAME", activation="selu"),
                                                keras.layers.Conv2DTranspose(1, kernel_size=3, strides=2, padding="SAME", activation="sigmoid"),
                                                keras.layers.Reshape([28, 28])])
    
        self.auto_encoder = keras.models.Sequential([self.encoder, self.decoder])
        self.auto_encoder.compile(loss="binary_crossentropy", optimizer=self.optimizer, metrics=[GeneralAE.rounded_accuracy])
```

We create an instance of `Conv2D_StandardAE` and start training.

```python
conv_standard_ae = Conv2D_StandardAE()
conv_standard_ae.build()
conv_standard_ae.train(X_train, X_train, epochs=20, validation_data=(X_valid, X_valid))
```

We show the reconstruction of some images in the valid set. As you can see, the quality of reconstructed images is generally better.

```python
conv_standard_ae.show_reconstructions(X_valid[:20], n_cols=5)
```

### <span style="color:#0b486b">II.4. Implement a denoising AutoEncoder</span> ###

Below is the implementation of the class `DenoisingAE`. We use the layer `keras.layers.GaussianNoise` to add Gaussian noises to the input.

```python
class DenoisingAE(GeneralAE):
    def __init__(self, optimizer = keras.optimizers.SGD(lr=0.1), noise= 0.2, seed= 6789):
        super(DenoisingAE, self).__init__(optimizer)
        self.noise = noise
        self.seed = seed
        tf.random.set_seed(self.seed)
        np.random.seed(self.seed)

    
    def build(self):
        self.encoder = keras.models.Sequential([keras.layers.Flatten(input_shape=[28, 28]),
                                                keras.layers.Dense(100, activation="selu"),
                                                keras.layers.GaussianNoise(self.noise),
                                                keras.layers.Dense(30, activation="selu")])
        
        self.decoder = keras.models.Sequential([keras.layers.Dense(100, activation="selu", input_shape=[30]),
                                                                  keras.layers.Dense(28 * 28, activation="sigmoid"),
                                                                  keras.layers.Reshape([28, 28])])
        
        self.auto_encoder = keras.models.Sequential([self.encoder, self.decoder])
        self.auto_encoder.compile(loss="binary_crossentropy", optimizer=self.optimizer, metrics=[GeneralAE.rounded_accuracy])
```

We create an instance of `DenoisingAE` and start training.

```python
denoise_ae = DenoisingAE()
denoise_ae.build()
denoise_ae.train(X_train, X_train, epochs=20, validation_data=(X_valid, X_valid))
```

We show the reconstruction of some images in the valid set. As you can see, it can denoise efficiently.

### <span style="color:#0b486b">II.5. Implement a sparse Auto-Encoder</span> ###

In what follows, we implement the class `SparseAE`. We declare a regularizer as a regularizer in `keras.regularizers` and apply it to the latent code.

```python
class SparseAE(GeneralAE):
    def __init__(self, optimizer = keras.optimizers.SGD(lr=0.1), regularizer = keras.regularizers.l1(l=0.01)):
        super(SparseAE, self).__init__(optimizer)
        self.regularizer = regularizer
    
    def build(self):
        self.encoder = keras.models.Sequential([keras.layers.Flatten(input_shape=[28, 28]),
                                                keras.layers.Dense(100, activation="selu"),
                                                keras.layers.Dense(30, activation="sigmoid", activity_regularizer=self.regularizer)])
        
        self.decoder = keras.models.Sequential([keras.layers.Dense(100, activation="selu", input_shape=[30]),
                                                                  keras.layers.Dense(28 * 28, activation="sigmoid"),
                                                                  keras.layers.Reshape([28, 28])])
        
        self.auto_encoder = keras.models.Sequential([self.encoder, self.decoder])
        self.auto_encoder.compile(loss="binary_crossentropy", optimizer=self.optimizer, metrics=[GeneralAE.rounded_accuracy])
```

We create an instance of `SparseAE` with L1 regularizer and start training.

```python
l1_sparse_ae = SparseAE(regularizer=keras.regularizers.l1(l=10))
l1_sparse_ae.build()
l1_sparse_ae.train(X_train, X_train, epochs=20, validation_data=(X_valid, X_valid))
```

We show the reconstruction of some images in the valid set. You can vary the regularizer parameter to see the effect to the reconstructed images.

For binary cross-entropy regularizer, we need to declare our own class `KLDivergenceRegularizer` inheritted from `keras.regularizers.Regularizer`.

```python
kl_divergence = keras.losses.kullback_leibler_divergence

class KLDivergenceRegularizer(keras.regularizers.Regularizer):
    def __init__(self, l, target=0.1):
        self.weight = l
        self.target = target
    def __call__(self, inputs):
        mean_activities = tf.reduce_mean(inputs, axis=0)
        return self.weight * (kl_divergence(self.target, mean_activities) + kl_divergence(1. - self.target, 1. - mean_activities))
```

We create an instance of `SparseAE` with binary cross-entropy regularizer and start training.

```python
kl_sparse_ae = SparseAE(regularizer=KLDivergenceRegularizer(l=0.01))
kl_sparse_ae.build()
kl_sparse_ae.train(X_train, X_train, epochs=20, validation_data=(X_valid, X_valid))
```

```python
def plot_percent_hist(ax, data, bins):
    counts, _ = np.histogram(data, bins=bins)
    widths = bins[1:] - bins[:-1]
    x = bins[:-1] + widths / 2
    ax.bar(x, counts / len(data), width=widths*0.8)
    ax.xaxis.set_ticks(bins)
    ax.yaxis.set_major_formatter(mpl.ticker.FuncFormatter(
        lambda y, position: "{}%".format(int(np.round(100 * y)))))
    ax.grid(True)
```

```python
def plot_activations_histogram(encoder, height=1, n_bins=10):
    X_valid_codings = encoder(X_valid).numpy()
    activation_means = X_valid_codings.mean(axis=0)
    mean = activation_means.mean()
    bins = np.linspace(0, 1, n_bins + 1)

    fig, [ax1, ax2] = plt.subplots(figsize=(10, 3), nrows=1, ncols=2, sharey=True)
    plot_percent_hist(ax1, X_valid_codings.ravel(), bins)
    ax1.plot([mean, mean], [0, height], "k--", label="Overall Mean = {:.2f}".format(mean))
    ax1.legend(loc="upper center", fontsize=14)
    ax1.set_xlabel("Activation")
    ax1.set_ylabel("% Activations")
    ax1.axis([0, 1, 0, height])
    plot_percent_hist(ax2, activation_means, bins)
    ax2.plot([mean, mean], [0, height], "k--")
    ax2.set_xlabel("Neuron Mean Activation")
    ax2.set_ylabel("% Neurons")
    ax2.axis([0, 1, 0, height])
```

#### <span style="color:#0b486b">II.6. Implement a Variational Auto-Encoder (VAE)</span> ####

We now implement a Variational Auto-Encoder (VAE) in the class `VariationalAE`. We start with the class `Sampling(keras.layers.Layer)` which serves as a stochastic layer.

```python
class Sampling(keras.layers.Layer):
    def call(self, inputs):
        mean, log_var = inputs
        return tf.random.normal(tf.shape(log_var)) * tf.math.exp(log_var / 2) + mean 
```

Here is the implementation of the class `VariationalAE`. Note that the final loss consists of two components as shown in the lecture.

```python
class VariationalAE(GeneralAE):
    def __init__(self, optimizer = keras.optimizers.SGD(lr=0.1), alpha= 0.5, seed= 6789):
        super(VariationalAE, self).__init__(optimizer)
        self.alpha = alpha
        self.seed = seed
        tf.random.set_seed(self.seed)
        np.random.seed(self.seed)
    
    def build(self):
        codings_size = 10
        inputs = keras.layers.Input(shape=[28, 28])
        z = keras.layers.Flatten()(inputs)
        z = keras.layers.Dense(150, activation="selu")(z)
        z = keras.layers.Dense(100, activation="selu")(z)
        codings_mean = keras.layers.Dense(codings_size)(z)
        codings_log_var = keras.layers.Dense(codings_size)(z)
        codings = Sampling()([codings_mean, codings_log_var])
        self.encoder = keras.models.Model(inputs=[inputs], outputs=[codings_mean, codings_log_var, codings])

        decoder_inputs = keras.layers.Input(shape=[codings_size])
        x = keras.layers.Dense(100, activation="selu")(decoder_inputs)
        x = keras.layers.Dense(150, activation="selu")(x)
        x = keras.layers.Dense(28 * 28, activation="sigmoid")(x)
        outputs = keras.layers.Reshape([28, 28])(x)
        self.decoder = keras.models.Model(inputs=[decoder_inputs], outputs=[outputs])
        
        _, _, codings = self.encoder(inputs)
        reconstructions = self.decoder(codings)
        self.auto_encoder = keras.models.Model(inputs=[inputs], outputs=[reconstructions])
        
        latent_loss = -self.alpha * tf.reduce_sum(1 + codings_log_var - tf.math.exp(codings_log_var) - tf.math.square(codings_mean),axis=-1)
        self.auto_encoder.add_loss(tf.math.reduce_mean(latent_loss) / 784.)
        self.auto_encoder.compile(loss="binary_crossentropy", optimizer=self.optimizer, metrics=[GeneralAE.rounded_accuracy])
```

We create an instance of `VariationalAE` and start training.

```python
vae = VariationalAE()
```

```python
vae.build()
```

```python
vae.train(X_train, X_train, epochs=20, validation_data=(X_valid, X_valid))
```

We show the reconstruction of some images in the valid set.

```python
vae.show_reconstructions(X_valid[:20], n_cols=5)
```

## <span style="color:#0b486b">III. Investigating and visualizing latent space</span> <span style="color:red">*****</span> ##

```python
from sklearn.manifold import TSNE
np.random.seed(6789)

def visualize_latent_codes(X=None, y=None, ae=None):
    tsne = TSNE()
    h_2D = ae.encode(X)
    h_2D = np.array(h_2D)
    if h_2D.ndim==3:
        h_2D = h_2D[0]
    h_2D = tsne.fit_transform(h_2D)
    h_2D = (h_2D - h_2D.min()) / (h_2D.max() - h_2D.min())
    y_unique = np.unique(y)
    colors = [('#27ae60', 'o'),
               ('#2980b9', 'o'),
               ('#8e44ad', 'o'),
               ('#f39c12', 'o'),
               ('#c0392b', 'o'),
               ('#87ae69', 'x'),
               ('#594019', 'x'),
               ('#5e11ad', 'x'),
               ('#d0e95b', 'x'),
               ('#e4fc17', 'x')]
    class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat", "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]

    for i in y_unique:
        hi = h_2D[np.where(y==i)]
        plt.scatter(hi[:, 0], hi[:, 1], label= class_names[i], c= colors[i][0], marker= colors[i][1],s=10, cmap="tab10")
        
    plt.legend(loc='upper left',ncol=10,fontsize=8)
    plt.axis("off")
    plt.show()
```

### <span style="color:#0b486b">IV.1. Tying encoder and decoder</span> ###

In auto-encoders, we can tie the weight matrices of encoders and decoders to reduce the number of parameters and take advantage from the benefit of sharing parameters. The following code shows you how to do that.

```python
class DenseTranspose(keras.layers.Layer):
    def __init__(self, dense, activation=None, **kwargs):
        self.dense = dense
        self.activation = keras.activations.get(activation)
        super().__init__(**kwargs)
    def build(self, batch_input_shape):
        self.biases = self.add_weight(name="bias",
                                      shape=[self.dense.input_shape[-1]],
                                      initializer="zeros")
        super().build(batch_input_shape)
    def call(self, inputs):
        z = tf.matmul(inputs, self.dense.weights[0], transpose_b=True)
        return self.activation(z + self.biases)
```

```python
class TyingAE(GeneralAE):
    def __init__(self, optimizer = keras.optimizers.SGD(lr=0.1)):
        super(TyingAE, self).__init__(optimizer)
    
    def build(self):
        dense1 = keras.layers.Dense(100, activation="selu")
        dense2 = keras.layers.Dense(30, activation="selu")
        self.encoder = keras.models.Sequential([keras.layers.Flatten(input_shape=[28, 28]),
                                                dense1,
                                                dense2])
        
        self.decoder = keras.models.Sequential([DenseTranspose(dense2, activation="selu"),
                                                DenseTranspose(dense1, activation="sigmoid"),
                                                keras.layers.Reshape([28, 28])])
        
        self.auto_encoder = keras.models.Sequential([self.encoder, self.decoder])
        self.auto_encoder.compile(loss="binary_crossentropy", optimizer=self.optimizer, metrics=[GeneralAE.rounded_accuracy])
```

### <span style="color:#0b486b">IV.2. RNN AutoEncoder</span> ###

Fortunately, RNN with memory cells (LSTM and GRU) can be used in encoder/decoder framework as shown in the following code.

```python
class RNN_AE(GeneralAE):
    def __init__(self, optimizer = keras.optimizers.SGD(lr=0.1)):
        super(RNN_AE, self).__init__(optimizer)
    
    def build(self):
        self.encoder = keras.models.Sequential([ keras.layers.LSTM(100, return_sequences=True, input_shape=[28, 28]),
                                                keras.layers.LSTM(30)])
        
        self.decoder = keras.models.Sequential([keras.layers.RepeatVector(28, input_shape=[30]),
                                                keras.layers.LSTM(100, return_sequences=True),
                                                keras.layers.TimeDistributed(keras.layers.Dense(28, activation="sigmoid"))])
        
        self.auto_encoder = keras.models.Sequential([self.encoder, self.decoder])
        self.auto_encoder.compile(loss="binary_crossentropy", optimizer=self.optimizer, metrics=[GeneralAE.rounded_accuracy])
```

```python
rnn_ae = RNN_AE()
rnn_ae.build()
rnn_ae.train(X_train, X_train, epochs=20, validation_data=(X_valid, X_valid))
```
